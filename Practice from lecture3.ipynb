```python
def sum_of_some_numbers(N):
    total = 0
    for i in range(1, N + 1):
        total += i
    return total

N = int(input("The input number: "))
print(f"The input number from 1 to ({N}) = {sum_of_some_numbers(N)}")
 #The input number from 1 to ({N}) ja icche bosate parbo but porer tuku {def er porer commant likha lagbe}
```

```python
#Write a function that checks whether a number is even two numbers
def Odd_or_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(Odd_or_even(10))
print(Odd_or_even(7))
```

```python
#Write a function that returns the larger number.
def x(a, b, c):
    if a >= b and a >= c:
        return "a is the largest number"
    elif b >= a and b >= c:
        return "b is the largest number"
    else:
        return "c is the largest number"

print(x(1, 2, 3))
print(x(5, 235, 34))
```
